```python
from pathlib import Path
from collections import defaultdict
import xml.etree.ElementTree as ET
from typing import Dict, Any

def parse_xbrl(filepath: str | Path) -> Dict[str, Any]:
    """
    Parse an XBRL (.xml) file and return a clean dictionary of tag names → values.
    
    Args:
        filepath: Path to the .xml file (string or Path object)
    
    Returns:
        Dictionary with clean tag names as keys and values (string or list if duplicates)
    
    Raises:
        FileNotFoundError: If file doesn't exist
        ET.ParseError: If XML is corrupted
    """
    filepath = Path(filepath)  # Convert string to Path if needed
    
    if not filepath.exists():
        raise FileNotFoundError(f"File not found: {filepath}")
    
    if not filepath.suffix.lower() == ".xml":
        print(f"Warning: File may not be XML: {filepath}")

    try:
        tree = ET.parse(filepath)
        root = tree.getroot()
    except ET.ParseError as e:
        raise ET.ParseError(f"Invalid XML file {filepath}: {e}")

    data = defaultdict(list)

    for elem in root.iter():
        # Clean tag name: remove namespace {http://...}
        tag_name = elem.tag.split("}")[-1] if "}" in elem.tag else elem.tag
        
        # Clean text value
        text = elem.text.strip() if elem.text else ""
        if text:
            data[tag_name].append(text)

    # Convert single-item lists to plain values
    result = {}
    for tag, values in data.items():
        result[tag] = values[0] if len(values) == 1 else values

    return result
```

```python
import json

def save_as_json(data_dict: dict, original_xml_path: Path | str):
    """
    Save dictionary as JSON with same name as original XML file but .json extension
    """
    xml_path = Path(original_xml_path)
    json_path = xml_path.with_suffix(".json")  # Changes .xml → .json
    
    with open(json_path, "w", encoding="utf-8") as f:
        json.dump(data_dict, f, indent=2, ensure_ascii=False)
    
    print(f"Saved → {json_path.name}")
```

```python
from pathlib import Path

folder = Path("../data")

all_financial_data = []

for xml_file in folder.glob("*.xml"):
    print(f"Processing: {xml_file.name}")
    financial_data = parse_xbrl(xml_file)
    save_as_json(financial_data, xml_file)

    all_financial_data.append(financial_data)
    
    # Now use it!
    print("Revenue:", financial_data.get("RevenueFromOperations", "Not found"))
    print("Quarter:", financial_data.get("ReportingQuarter", "Not found"))
    print("-" * 50)
```

```python
import os
from pathlib import Path

print("JSON Files and Their Sizes:")
print("-" * 50)

total_size = 0

json_files = list(Path("../data").glob("*.json"))

if not json_files:
    print("No JSON files found!")
else:
    for file_path in json_files:
        size_bytes = file_path.stat().st_size
        size_kb = size_bytes / 1024
        
        print(f"{file_path.name:40} → {size_bytes:,} bytes ({size_kb:.1f} KB)")
        total_size += size_bytes

print("-" * 50)
print(f"Total JSON files: {len(json_files)}")
print(f"Total size: {total_size:,} bytes ({total_size/1024:.1f} KB)")
```
